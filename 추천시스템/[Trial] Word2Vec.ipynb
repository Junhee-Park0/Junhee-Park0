```python
hashtags.head()
only_tags = hashtags['tag'].to_list()
len(only_tags)
```

```python
# 1. "#" 제거 - 모델 input
import re
tags_wo_hash = []
for tag in only_tags:
  tags_wo_hash.append(re.sub(r"[#\[\]\'\,]","",tag).split())
tags_wo_hash[:2]
```

굳이 형태소 분리를 안 해도 되긴 하나보네.. 이미 단어 단위라 그냥 들어가는 건가

```python
# 2-2. 형태소 분리 o - 그냥 한 글자별로 분리해버린다.. 문장이 아니라서 제대로 인식을 못하는 것 같아 안 하는 게 낫겠다
def tokenize_tags(tags):
  tokenized_tags = []

  for tag in tags:
    tokenized = mecab.morphs(tag)
    tokenized_tags.append(tokenized)

  return tokenized_tags

tokenize_tags(tags_wo_hash[0])
```

```python
# 2-1. 형태소 분리 x

## CBOW : 주변으로 중심 예측
## Skip-Gram : 중심으로 주변 예측
## min_count 일단은 2로
model = Word2Vec(sentences = tags_wo_hash, vector_size = 50, window = 3, min_count = 2, sg = 1, workers = 4)
```

### 4. 평균 구하는 함수 생성

```python
def avg_vector(tag):
  tag_vector = model.wv[tag]
  avg_vector = sum(tag_vector)/len(tag_vector)
  return avg_vector
```

```python
new_tag = ['가을','사랑','연인','데이트'] # 몽글은 단어셋에 없음...
avg_vector(new_tag)
```

새로운 단어가 들어오면 어떻게 처리하지? \\
1. 그냥 0벡터 처리해서 없애
2. 유사 단어 찾기..?

```python
# 1. 그냥 무시.. 없애
def avg_vector_zero(tag):
  tag_vector = []
  for word in tag:
    if word in model.wv:
      tag_vector.append(model.wv[word])
    else: # 학습한 단어에 없으면 그냥 패스
      continue

  avg_vector = sum(tag_vector)/len(tag_vector)
  return avg_vector
```

```python
zeros_index = {339,390,896}
tags_wo_hash_wo_zero = [element for i, element in enumerate(tags_wo_hash) if i not in zeros_index]
len(tags_wo_hash_wo_zero)
```

```python
original_vectors = [avg_vector_zero(tag) for tag in tags_wo_hash_wo_zero]
```

### 6. 코사인 유사도

```python
# original_vectors : 기존 태그들의 임베딩 결과 벡터
# 새로운 태그에 avg_vector_zero 적용 : 새로운 태그의 임베딩 결과 벡터
new_tag = ['가을','사랑','연인','몽글','데이트']
new_vector = avg_vector_zero(new_tag)
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cos_sim(A, B):
  return dot(A, B)/(norm(A)*norm(B))
```

```python
print("첫 해시태그와의 유사도:", cos_sim(original_vectors[0], new_vector))
```

### 7. 최종) 유사한 태그 출력

```python
from heapq import nlargest
from heapq import nsmallest

similarities = []
for i in range(len(original_vectors)):
  similarities.append(cos_sim(original_vectors[i], new_vector))

top_5 = [idx for idx,_ in nlargest(5, enumerate(similarities), key=lambda x:x[1])]
top_5_hashtags = [tags_wo_hash_wo_zero[i] for i in top_5]
```

```python
print(f"{new_tag}와 비슷한 해시태그들: \n {top_5_hashtags[0]} \n {top_5_hashtags[1]} \n {top_5_hashtags[2]} \n {top_5_hashtags[3]} \n {top_5_hashtags[4]}")
```

```python
from heapq import nsmallest

least_5 = [idx for idx,_ in nsmallest(5, enumerate(similarities), key=lambda x:x[1])]
least_5_hashtags = [tags_wo_hash_wo_zero[i] for i in top_5]
```

```python
print(f"{new_tag}와 먼 해시태그들: \n {least_5_hashtags[0]} \n {least_5_hashtags[1]} \n {least_5_hashtags[2]} \n {least_5_hashtags[3]} \n {least_5_hashtags[4]}")
```
